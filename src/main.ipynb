## Preprocessing the Data

```python
train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)
train_data["population"] = np.log(train_data["population"] + 1)
train_data["households"] = np.log(train_data["households"] + 1)
```

```python
train_data.hist(figsize=(16,8))
```

```python
dis_train_data = train_data.drop('ocean_proximity', axis=1)
correlation_matrix = dis_train_data.corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(dis_train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
# While building a model its hard to work with non-numerical data.
# So we need to convert them into numerical values.
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity)
```

```python
# train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop('ocean_proximity', axis=1)

# Create one-hot encoded columns for 'ocean_proximity'
dummies = pd.get_dummies(train_data['ocean_proximity'])

# Replace True/False with 1/0
dummies = dummies.astype(int)

# Join the one-hot encoded columns to your DataFrame
train_data = train_data.join(dummies)

# Drop the original 'ocean_proximity' column
train_data = train_data.drop('ocean_proximity', axis=1)
```
